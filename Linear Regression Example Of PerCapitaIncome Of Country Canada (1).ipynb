# DataSet: Canada Adjusted Net National Income Per Capita VS Year Here, we are going to predict the net income by Linear Regression and Plot them

```python
# importing required libraries
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
# reading the data by pandas as read_csv
df = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/Exercise/canada_per_capita_income.csv')
df
```

```python
# changing the column name
df_new = df.rename(columns={'per capita income (US$)' : 'PerCapitaIncome'})
```

```python
# finding the first five rows
df_new.head()
```

```python
#plotting the data
%matplotlib inline
plt.xlabel('year')
plt.ylabel('PerCapitaIncome')
plt.scatter(df_new.year,df_new.PerCapitaIncome,color='red',marker='+')
```

```python
# dividing the columns by drop
new_df_new = df_new.drop('PerCapitaIncome',axis = 'columns')
new_df_new
```

```python
# dividing percapita income column:
PerCapitaIncome = df_new.PerCapitaIncome
PerCapitaIncome
```

```python
# Training& fitting the Linear Regression model
model = linear_model.LinearRegression()
model.fit(new_df_new,PerCapitaIncome)
```

```python
#predicting the model
model.predict([[2020]])
```

```python
# slope/weight
model.coef_
```

```python
# bias
model.intercept_
```

```python
2020*828.46507522 + (-1632210.7578554575)
```

```python
model.predict([[2025]])
```

```python
# seperating years column:
D = pd.read_csv('Years.csv')
D.head()
```

```python
# predicting the outcomes:
model.predict(D)
```
